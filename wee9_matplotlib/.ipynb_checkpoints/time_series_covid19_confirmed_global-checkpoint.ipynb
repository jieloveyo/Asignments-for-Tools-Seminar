```python
%matplotlib inline 
import matplotlib.pyplot as plt
from datetime import datetime  # 将字符串转换为对应的日期格式
import csv
import pandas as pd
import numpy as np
```

```python
CSV_FILE_PATH = 'time_series_covid19_confirmed_global.csv'
df = pd.read_csv(CSV_FILE_PATH)
df
```

####  删除某一行或某一列
* df.drop(["列名"]，axis=1)
* df.drop(columns=["列名"])
* df.drop([0,1])  删除索引为0、1的整行

```python
df.drop(["Country/Region"],axis=1,inplace=True)
```

```python
df.replace(" ?", np.nan, inplace=True)
df.isna().sum()
```

```python
test = df.sum(axis=0)
```

```python
type(test)
```

```python
test
```

```python
#画图
test.plot(figsize=(30,20),kind='bar')

#字体大小
plt.title('confirmed_people_sum_bar',fontdict={'weight':'normal','size': 30}) #改变图标题字体
plt.xlabel('Time', fontdict={'weight': 'normal', 'size': 13})#改变坐标轴标题字体
plt.ylabel('sum',  fontdict={'weight': 'normal', 'size': 17})
#坐标的旋转率
plt.xticks(rotation=40)

#设置x轴/y轴的范围


#保存图片
plt.savefig('times_confirmed_sum_bar.pdf')
```

```python
#画图
test.plot(figsize=(15,8),kind='line',linewidth=4,color='red')

#字体大小
plt.title('confirmed_people_sum_bar',fontdict={'weight':'normal','size': 15}) #改变图标题字体
plt.xlabel('Time', fontdict={'weight': 'normal', 'size': 13})#改变坐标轴标题字体
plt.ylabel('sum',  fontdict={'weight': 'normal', 'size': 17})
#坐标的旋转率
plt.xticks(rotation=40)

#设置x轴/y轴的范围
plt.ylim(0,2500000)

#保存图片
plt.savefig('times_confirmed_sum_line.pdf')
```

```python
#画图
test.plot(figsize=(20,18),kind='barh',linewidth=4,color='red')

#字体大小
plt.title('confirmed_people_sum_bar',fontdict={'weight':'normal','size': 15}) #改变图标题字体
plt.xlabel('Sum', fontdict={'weight': 'normal', 'size': 13})#改变坐标轴标题字体
plt.ylabel('Time',  fontdict={'weight': 'normal', 'size': 17})
#坐标的旋转率
plt.xticks(rotation=40)

#保存图片
plt.savefig('times_confirmed_sum_barh.pdf')
```
