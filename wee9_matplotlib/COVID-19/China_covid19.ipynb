```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Confirmed = pd.read_csv("csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv")
Death = pd.read_csv("csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv")
Recovered = pd.read_csv("csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv")
```

```python
Confirmed.head()
```

```python
drop_attr = ['Lat','Long']
Confirmed.drop(drop_attr,axis=1,inplace=True)
Death.drop(drop_attr,axis=1,inplace=True)
Recovered.drop(drop_attr,axis=1,inplace=True)
```

```python
China_confirmed = Confirmed[Confirmed["Country/Region"] == 'China']
China_death = Death[Death["Country/Region"] == 'China']
China_recovered = Recovered[Recovered["Country/Region"] == 'China']
```

```python
China_confirmed.head()
```

```python
drop_attr = ['Province/State','Country/Region']
China_confirmed.drop(drop_attr,axis=1,inplace=True)
China_recovered.drop(drop_attr,axis=1,inplace=True)
China_death.drop(drop_attr,axis=1,inplace=True)
```

```python
China_confirmed.head()
```

```python
y_confirmed = China_confirmed.sum(axis=0)
y_recovered = China_recovered.sum(axis=0)
y_death = China_death.sum(axis=0)
```

```python
y_death
```

```python
type(y_death)
```

```python
y_confirmed.plot(label='y_confirmed',linewidth=2,c='r')
y_recovered.plot(label='y_recovered',linewidth=2,c='g')
y_death.plot(label='y_death',linewidth=2)
plt.legend()
plt.savefig('China_line.pdf')
```
