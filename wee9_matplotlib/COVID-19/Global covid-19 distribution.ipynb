```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Confirmed = pd.read_csv("csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv")
```

```python
Confirmed.head()
```

```python
drop_attr = ['Lat','Long','Province/State']
Confirmed.drop(drop_attr,axis=1,inplace=True)
```

```python
Confirmed.head()
```

```python
Total_confirmed = Confirmed.groupby("Country/Region").sum()
```

```python
Total_confirmed
```

```python
data = Total_confirmed['4/17/20']
data
```

```python
type(data)
```

```python
t_data = data[data>40000]
```

```python
other = data[data<40000].sum()
other
```

```python
t_data['Other'] = other
```

```python
t_data
```

```python
# 设置x，y轴刻度一致，这样饼图才能是圆的
plt.axis('equal')

t_data.plot.pie(figsize=(12,12),labeldistance = 1.1,autopct = '%3.1f%%',shadow = False,startangle = 90,pctdistance = 0.6, title='Global covid-19 distribution')

plt.legend()
#labeldistance，文本的位置离远点有多远，1.1指1.1倍半径的位置
#autopct，圆里面的文本格式，%3.1f%%表示小数有三位，整数有一位的浮点数
#shadow，饼是否有阴影
#startangle，起始角度，0，表示从0开始逆时针转，为第一块。一般选择从90度开始比较好看
#pctdistance，百分比的text离圆心的距离
#patches, l_texts, p_texts，为了得到饼图的返回值，p_texts饼图内部文本的，l_texts饼图外label的文本

plt.savefig('Global covid-19 distribution.pdf')
```
